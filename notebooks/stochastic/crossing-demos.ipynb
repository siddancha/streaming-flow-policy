```python
import matplotlib.pyplot as plt
import numpy as np
from flow_policy.stochastic_flow_policy import StochasticFlowPolicy
from flow_policy.plot_stochastic import (
    plot_probability_density_q,
    plot_probability_density_ε,
    plot_probability_density_with_trajectories_q,
    plot_probability_density_with_trajectories_ε,
)

from pydrake.all import (
    PiecewisePolynomial,
    Trajectory,
)

# Set seed
np.random.seed(0)
```

## Set sigma parameters

```python
σ0 = 0.005
σ1 = 0.05
```

```python
def demonstration_traj_right() -> Trajectory:
    return PiecewisePolynomial.CubicWithContinuousSecondDerivatives(
        breaks=[0.00, 0.25, 0.50, 0.75, 1.0],
        samples=[[0.00, 0.75, 0.00, -0.75, 0.00]],
        sample_dot_at_start=[[3.0]],
        sample_dot_at_end=[[3.0]],
    )

def demonstration_traj_left() -> Trajectory:
    return PiecewisePolynomial.CubicWithContinuousSecondDerivatives(
        breaks=[0.00, 0.25, 0.50, 0.75, 1.0],
        samples=[[0.00, -0.75, 0.00, 0.75, 0.00]],
        sample_dot_at_start=[[-3.0]],
        sample_dot_at_end=[[-3.0]],
    )

traj_right = demonstration_traj_right()
traj_left = demonstration_traj_left()
```

## Plot demonstration trajectories

```python
"""
Plot demonstration trajectory on x-y plane where x axis is the state in [-1, 1]
and y axis is the time in [0, 1].
"""
times = np.linspace(0, 1, 100)
plt.plot(traj_right.vector_values(times)[0], times, color='blue', alpha=0.9)
plt.plot(traj_left.vector_values(times)[0], times, color='red', alpha=0.9)
plt.xlim(-1, 1)
plt.ylim(0, 1)
plt.xlabel('Configuration')
plt.ylabel('Time')
plt.title('Demonstration Trajectories')
plt.grid(True)
plt.show()
```

## Conditional flow

```python
fp = StochasticFlowPolicy(trajectories=[traj_right], prior=[1.0], σ0=σ0, σ1=σ1)
```

## Plot conditional probability path of right trajectory

```python
fig = plt.figure(figsize=(8, 4), dpi=150)

xs = np.linspace(-1, 1, 200)
ts = np.linspace(0, 1, 200)

gs = fig.add_gridspec(1, 2, width_ratios=[1, 1])
ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1])

plot_probability_density_q(fp, ts, xs, ax1)
plot_probability_density_ε(fp, ts, xs, ax2)

ax1.set_title('Configuration (q) Probability Density', size='large')
ax2.set_title('Epsilon (ε) Probability Density', size='large')

ax1.set_xlabel('Configuration (q)')
ax1.set_ylabel('Time ⟶')
ax2.set_xlabel('Epsilon (ε)')
ax2.set_ylabel('Time ⟶')

plt.tight_layout()
plt.show()
```

## Plot trajectories under conditional flow of right trajectory

```python
q_starts = [0.0] * 10
ε_starts_pos = np.abs(np.random.randn(5))
ε_starts_neg = -np.abs(np.random.randn(5))
ε_starts = sorted(np.concatenate([ε_starts_neg, ε_starts_pos]))
colors = ['red'] * 5 + ['red'] * 5

fig = plt.figure(figsize=(8, 4), dpi=150)

gs = fig.add_gridspec(1, 2, width_ratios=[1, 1])
ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1])

plot_probability_density_with_trajectories_q(
    fp, ax1, q_starts, ε_starts, colors, num_points_x=400)
plot_probability_density_with_trajectories_ε(
    fp, ax2, q_starts, ε_starts, colors, num_points_x=400)
plt.show()
```

# Marginal flow

```python
fp = StochasticFlowPolicy(trajectories=[traj_right, traj_left], prior=[0.5, 0.5], σ0=σ0, σ1=σ1)
```

## Plot marginal probability path

```python
fig = plt.figure(figsize=(8, 4), dpi=150)

xs = np.linspace(-1, 1, 200)
ts = np.linspace(0, 1, 400)

gs = fig.add_gridspec(1, 2, width_ratios=[1, 1])
ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1])

plot_probability_density_q(fp, ts, xs, ax1)
plot_probability_density_ε(fp, ts, xs, ax2)

ax1.set_title('Configuration (q) Probability Density', size='large')
ax2.set_title('Epsilon (ε) Probability Density', size='large')

ax1.set_xlabel('Configuration (q)')
ax1.set_ylabel('Time ⟶')
ax2.set_xlabel('Epsilon (ε)')
ax2.set_ylabel('Time ⟶')

plt.tight_layout()
plt.show()
```

## Plot trajectories under marginal flow

```python
q_starts = [0.0] * 20
ε_starts_pos = np.abs(np.random.randn(10))
ε_starts_neg = -np.abs(np.random.randn(10))
ε_starts = sorted(np.concatenate([ε_starts_pos, ε_starts_neg]))
colors = ['blue'] * 10 + ['red'] * 10

fig = plt.figure(figsize=(8, 4), dpi=150)

gs = fig.add_gridspec(1, 2, width_ratios=[1, 1])
ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1])

plot_probability_density_with_trajectories_q(
    fp, ax1, q_starts, ε_starts, colors, heatmap_alpha=0.5, linewidth=1, num_points_x=400,
)
plot_probability_density_with_trajectories_ε(
    fp, ax2, q_starts, ε_starts, colors, heatmap_alpha=0.5, linewidth=2, num_points_x=400,
)
plt.show()
```
